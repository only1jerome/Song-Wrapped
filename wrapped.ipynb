```python
import pandas as pd
sd = pd.read_json('StreamingHistory_music_0.json')
sd
```

```python
# Function to convert msPlayed to MM:SS format
def convert_ms_to_minutes_seconds(ms):
    minutes = ms // 60000 
    seconds = (ms % 60000) // 1000  
    return f"{minutes}:{seconds:02d}"  


sd['timePlayed'] = sd['msPlayed'].apply(convert_ms_to_minutes_seconds)
sd
```

```python
artist_time = sd.groupby('artistName')['msPlayed'].sum().reset_index()
artist_time = artist_time.sort_values('msPlayed', ascending=False).head(20)
artist_time['TotalMinutesPlayed'] = artist_time['msPlayed'].apply(convert_ms_to_minutes_seconds)
artist_time.drop(columns=['msPlayed'], inplace=True)
artist_time
```

```python
song_time = sd.groupby(['artistName', 'trackName'])['msPlayed'].sum().reset_index()
song_time = song_time.sort_values('msPlayed', ascending=False).head(20)
song_time['TotalMinutesPlayed'] = song_time['msPlayed'].apply(convert_ms_to_minutes_seconds)
song_time.drop(columns=['msPlayed'], inplace=True)
song_time
```
